```python
# CITIRE CSV UTILIZAND PYSPARK

df_spark = spark.createDataFrame(pd.read_csv('/content/drive/MyDrive/Colab Notebooks/proiect-bda/proiect_bda_news2.csv'))
df_spark

# AFISARE TABEL
df_spark.show(100)

# AFISARE SCHEMA
df_spark.printSchema()
```

```python
# REDENUMIRE COLOANA SOURCE.NAME in sourceName
# ELIMINA COLOANA SOURCE.ID
# SCHIMBA FORMATUL DATEI LA YYYY-MM-DD SI SUPRASCRIE IN ACEEASI COLOANA

from pyspark.sql import functions as F

df_spark2 = df_spark.withColumnsRenamed({'source.name': 'sourceName'}).drop('source.id')
df_spark2 = df_spark2.withColumn('publishedAt', F.date_format(F.col('publishedAt'), 'yyyy-MM-dd'))
df_spark2.show()

# df_spark.show()
```

```python
# NR DATE INTRARE

df_spark2.count()
```

```python
df_spark2.toPandas()
```

```python
# VERIFICARE VALORI NULE (NAN)

df_spark2.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df_spark2.columns]).show()
```

```python
# AFISARE RANDURI CARE NU CONTIN VALORI NULE

from pyspark.sql.functions import isnan
from functools import reduce

# IDENTIFICA SI ELIMINA RANDURILE CARE CONTIN VALORI NaN IN ORICE COLOANA
df_spark3 = df_spark2.filter(~reduce(lambda x, y: x | y, (isnan(c) for c in df_spark2.columns))).na.drop()

df_spark3.show()
df_spark3.count()
```

```python
# NUMARARE ARTICOLE PENTRU FIECARE SURSA

df_spark3.groupBy('sourceName').count().show(100)
```

```python
# NUMARARE ARTICOLE PENTRU FIECARE ZI

df_spark3.groupBy('publishedAt').count().show(200)
```

```python
# AFISARE GRAFIC NR ARTICOLE PE SURSA
import pyspark.pandas as ps
# import matplotlib.pyplot as plt
%matplotlib inline

nr_articole_sursa = df_spark3.groupBy('sourceName').count().collect()

source = [row[0] for row in nr_articole_sursa]
values = [row[1] for row in nr_articole_sursa]

# df = ps.DataFrame( {'values': values}, index=source)
# df.plot.bar(x='values')

# df = ps.DataFrame( {'values': values}, index=source)
# df.plot.bar(y='values')

df = ps.DataFrame({'label': source, 'values': values})

df.plot.bar(x='label', y='values')


```

```python
# AFISARE GRAFIC NR ARTICOLE PE ZI

import pyspark.pandas as ps
# import matplotlib.pyplot as plt

nr_articole_zi = df_spark3.groupBy('publishedAt').count().collect()

day = [row[0] for row in nr_articole_zi]
values = [row[1] for row in nr_articole_zi]

df = ps.DataFrame( {'Values': values, 'Day': day})
fig = df.plot.bar(x='Day', y='Values')

fig.show()
```
